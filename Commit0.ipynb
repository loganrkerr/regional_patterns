That said, the broader patterns that emerge from this sort of analysis are still useful in comparing the startup scenes in each city.

##Step-by-step walkthrough of methods (using SF as example)
--------------------------------

I based my analysis on market tags. AngelList allows startups to enter multiple tags that describe the markets or spaces they are targeting. These are entered by humans, so a certain degree of error/redundancy/subjectivity should be expected. Furthermore, the tags are nested in a tree structure but our analysis assumes a nonhierarchical network. As a result, our analysis ignores the implicit relationship between a tag and the tags it subsets or is subsetted by (e.g. the 'Consumer' tag is a parent of 'Consumer Internet', but the set of startups categorized as 'Consumer' won't necessarily include startups that are categorized as 'Consumer Internet'). That said, in aggregate I believe these tags provide useful information regarding the startup ecosystem.

The following snippet of code transforms the json-like objects returned by AngelList API to a sparse matrix. 

```python
def sparse_encoding(items):
    dictionary = {}
    maxid = 0
    ii = []
    jj = []
    v = []
    for i,item in enumerate(items):
        for tag in item:
            j = dictionary.setdefault(tag,maxid)
            if j==maxid:
                maxid = maxid+1
            jj.append(j)
            ii.append(i)
            v.append(1)
            
    return coo_matrix((v,(ii,jj))),dictionary
```

The result is a sparse matrix where there's non-zero element at position (i,j) if startup i has the market tag j. There are about 9k startups and some 500 market tags in the dataset. Note that that's probably not all startups in SF, only ones which have a profile on AngelList with a correct location tag.

Below is a sparsity pattern representation of our matrix. As above, each blue dot is a non-zero element with position given by (i,j).

```python
data,dictionary = sparse_encoding([set([market['id'] for market in startup['markets']]) for startup in non_hidden_sf])
plt.spy(data.T)
```

Of course, this graph doesn't give us much insight into the relationships between startups and markets. To analyse this data by relationship, we compute pairwise affinities between startups. Here, it makes sense to use cosine distance between the binary vectors describing the market tags associated with startups. 

Whereas our earlier graph depicts occurrence of startup-market relationship, the graph below visualizes the similarity or "distance" between startups based on their shared market relationships.

```python
A = 1-squareform(pdist(data.todense(),'cosine'))
A[np.isnan(A)] = 0
plt.spy(A)
plt.xlabel('startups')
plt.ylabel('startups')
plt.title('similarity between SF startups')
```

The result is a sparse matrix. The structure of this matrix is mostly hidden because the startups are ordered based on number of followers on AngelList (i.e. randomly). We'll need to reorder the startups so that more structure becomes visible. To do this, we can perform simple Laplacian eigenmaps: compute the graph laplacian L, and compute it's smallest eigenvectors. The second smallest eigenvector will be a good dimension  along which to order startups (for more info on Laplacian eigenmaps, check out [the original paper](http://www.cse.ohio-state.edu/~mbelkin/papers/LEM_NC_03.pdf)).

```python
Areg = A + 0.02*np.ones(A.shape)
D = np.diag(sum(Areg,0))
L = D - Areg
u,v = eigs(L,3,D,which='SM')
```

```python
order = sorted(range(data.shape[0]),key=lambda(x):v[x,1])
plt.spy(A[order,:][:,order])
```

So now that the startups are ordered by the second smallest eigenvalue of the Laplacian, a cluster structure becomes clearly visible. Let's look at it in more detail, using the second and third smallest eigenvalues to create a map of startups. In the figure below the x axis is the second, y axis is the third smallest eigenvector of the Laplacian, but the meaning of the coordinates is not really relevant. Each dot represents a startup. Startups that are more similar to each other (in terms of cosine distance between their market tags) will be closer to each other on the plot.

```python
plt.plot(v[:,1],v[:,2],'.')
```

A number of clusters emerge. By running clustering on the data, we can identify these clusters. 

```python
clustering = SpectralClustering(affinity='precomputed',n_clusters = 7)
labels = clustering.fit_predict(Areg)
plt.scatter(v[:,1],v[:,2],c=labels)
```

When we colour the dots according to the cluster labels spectral clustering found, we see that it roughly identified the clusters that were visible anyway.  It does not look perfect, but that's because of the imperfect visualisation of the graph: Laplacian eigenmaps only gives us an approximate look at the structure of the data. 

Below is a list the clusters that spectral clustering found, along with the 3 most popular market tags in each cluster and 10 examples of startups for each.

```python
for l in range(7):
    cnt = Counter([market['name'] for startup,label in zip(non_hidden_sf,labels) if label==l for market in startup['markets']])
    print 'Cluster %d, (%s)'%(l,', '.join([item[0] for item in cnt.most_common(3)]))
    for name,markets in [(startup['name'],[market['name'] for market in startup['markets']]) for startup,label in zip(non_hidden_sf,labels) if label==l][:10]:
        print '\t%s:\t\t%s'%(name,markets)
```

- `SaaS, enterprise software, B2B`

Glancing over our lists, the distribution and groupings of tags seems to make sense. `e-commerce`, `marketplaces`, `fashion`, and `retail` often appear together and are among the most prominent clusters in each city. Similarly, `SaaS`, `enterprise software`, and `B2B` crop up in the same clusters, as do `health care`, and `health and wellness`. In turn, LA is the only city to feature a `music`, `entertainment industry`, and `film` cluster, and edtech features heavily in Boston's startup ecosystem. 

Looking at these ecosystems more critically, what struck me is how our data seems to overturn many of the commonly held assumptions regarding both region-agnostic trends and city-specific stereotypes. Despite its traditional association as being a hub for biotech, Boston's health tech scene comes in second to New York's. Similarly, for all the hype about the fintech and health IT trends, London is the only city with a pure `financial services`, `finance technology`, and `finance` cluster and `health care information technology` only shows up among the Boston and SF clusters. 

Also surprising is the apparent absence of `big data` and `cloud computing`. Despite the media's attention towards these markets, SF is the only ecosystem in which they appear to be significant (and even within that ecosystem, `big data` and `cloud computing` occur separately rather than together as part of a pure `big data` or `cloud computing` cluster). One explanation for `cloud computing`'s rarity could be that the market is mostly dominated by big players (Amazon, Microsoft, Google, etc.) and our analysis doesn't take into account the size of a startup/company. That said, this still doesn't explain why `big data` appears so infrequently. 

Finally, I just want to touch on the `social commerce` tag. I'm a big believer in [market networks](http://techcrunch.com/2015/06/27/from-social-to-market-networks/) (all credit to James Currier for coining the term), which are best approximated by AngelList's `social commerce` and `collaborative consumption` tags and can be identified by the way they:
- Combine the main elements of both networks and marketplaces
- Use SaaS workflow software to focus action around longer-term projects, not just a quick transaction
- Promote the service provider as a differentiated individual, helping to build long-term relationships

According to James, market networks represent an evolution of the social network trend, which makes it significant that the `social commerce` tag only appears in the SF and LA clusters. I plan on repeating this analysis again in a few months to see if the tags and clusters have changed, and I'll be interested to see if `social commerce` or the other descriptions of market networks feature more prominently and more broadly as time goes on.
